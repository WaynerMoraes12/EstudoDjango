```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

altura, largura = 400, 800


escocia = np.zeros((altura, largura, 3), dtype=np.uint8)
escocia[:] = (255, 0, 0)  # azul
cv2.line(escocia, (0, 0), (largura, altura), (255, 255, 255), 80)
cv2.line(escocia, (0, altura), (largura, 0), (255, 255, 255), 80)

cv2_imshow(escocia)


inglaterra = np.zeros((altura, largura, 3), dtype=np.uint8)
inglaterra[:] = (255, 255, 255)
cv2.line(inglaterra, (largura//2, 0), (largura//2, altura), (0, 0, 255), 100)
cv2.line(inglaterra, (0, altura//2), (largura, altura//2), (0, 0, 255), 100)

cv2_imshow(inglaterra)


irlandaN = np.zeros((altura, largura, 3), dtype=np.uint8)
irlandaN[:] = (255, 255, 255)
cv2.line(irlandaN, (0, 0), (largura, altura), (0, 0, 255), 60)
cv2.line(irlandaN, (0, altura), (largura, 0), (0, 0, 255), 60)

cv2_imshow(irlandaN)


soma1 = cv2.addWeighted(escocia, 0.5, inglaterra, 0.5, 0)

reino_unido = cv2.addWeighted(soma1, 0.7, irlandaN, 0.3, 0)

cv2_imshow(reino_unido)

```
